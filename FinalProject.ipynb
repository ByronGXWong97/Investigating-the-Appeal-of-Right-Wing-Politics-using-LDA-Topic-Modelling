## Summary Statistics: Words with highest TF-IDF scores

TF-IDF is a statistic that measures the importance of a word to a collection of documents. One measure of how important a word may be is its term frequency (tf), how frequently a word occurs in a document. Another approach is to look at a term’s inverse document frequency (idf), which decreases the weight for commonly used words and increases the weight for words that are not used very much in a collection of documents. This can be combined with term frequency to calculate a term’s tf-idf (the two quantities multiplied together), the frequency of a term adjusted for how rarely it is used.

```python
vect1 = TfidfVectorizer(stop_words="english")
sparse_tfidf = vect1.fit_transform(Trumpcorpus)
tfidf = pd.DataFrame(sparse_tfidf.toarray(), columns=vect1.get_feature_names())
tfidf.max().sort_values(ascending=False).head(20)

#This identifies the 20 words with highest tf-idf weights.
#We start to see some interesting keywords. 
#Next, we conduct topic modeling to understand the context in which these words are used.
```

## Topic modeling with Latent Dirichlet Allocation (LDA)

```python
from sklearn.decomposition import LatentDirichletAllocation
n_topics = 5
vect3 = CountVectorizer(max_df=0.7, min_df=0.2,stop_words='english')
tf = vect3.fit_transform(Trumpcorpus)
vocab=vect3.get_feature_names()
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=20, random_state=0)
lda = lda.fit(tf)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #{}:".format(topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
print()
```

```python
tf_feature_names = vect3.get_feature_names()
print_top_words(lda, tf_feature_names, 10)
```

We have identified 5 key topics from Trump's corpus, which are pretty informative about the key narratives used by Trump to shore up support for his campaign. 

Topic 0 is centered on US foreign policy decisions. During Trump's campaign, he denounced Obama's foreign policy decisions. He shifted the US embassy from Tel Aviv to Jerusalem (he promised to be the best "friend" to "israel"), and slammed Obama's nuclear deal with "iran". These rhetorics give us a hint of Trump's leanings in partisan politics.

In Topic 2, we see how Trump accuses his political opponent Clinton of corruption. "speeches" could allude to the fact that Clinton and Bill made over $153 million giving speeches to lobbyists, CEOs and foreign governments since 2001. "fbi" and "secret" could allude to Clinton's use of a private email server during her time as the Secretary of State, compromising national security. Trump also accused the "justice" "department" of colluding with Clinton. These inferences are corroborated by Greven (2016), where he argues that "populism's central and permanent narrative is the juxtaposition of a corrupt political class, elite or establishment"

Topic 3 could possibly be alluding to the 2011 intervention in Libya that his "opponent" Clinton was involved in, leaving behind a "failed" state in which "isis" militants could unleash "violence" and grow in strength. Fuchs (2017) further suggests that such scapegoating is part of Trump's ideological campaign.

In Topic 4, we see how Trump mobilises ethnocentric and nativist sentiments to appeal to White nationalists. He does so by creating an antagonism of "us versus them" (Greven 2016). Trump proposed the construction of a wall along the "border" with "mexico" to halt the flow of immigrants into the US. He also argued for the 11 million illegal immigrants in the US to be deported (many of whom are from Mexico). Fuchs (2017) argues that this construction of out-groups (presented as threats to the greatness of the America) is a feature of Trump's authoritarianism.

## Plot word clouds of topics

```python
#To make our results more visually intuitive, we generate a word cloud for each topic.

import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud

first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
fifth_topic = lda.components_[4]

first_topic_words = [tf_feature_names[i] for i in first_topic.argsort()[:-25 - 1 :-1]]
second_topic_words = [tf_feature_names[i] for i in second_topic.argsort()[:-25 - 1 :-1]]
third_topic_words = [tf_feature_names[i] for i in third_topic.argsort()[:-25 - 1 :-1]]
fourth_topic_words = [tf_feature_names[i] for i in fourth_topic.argsort()[:-25 - 1 :-1]]
fifth_topic_words = [tf_feature_names[i] for i in fifth_topic.argsort()[:-25 - 1 :-1]]

topics_combi=[first_topic_words,second_topic_words,third_topic_words,fourth_topic_words,fifth_topic_words]

for i in topics_combi:
    cloud = WordCloud(stopwords='english',
                          background_color='white',
                          width=2500,
                          height=1800).generate(" ".join(i))
    plt.imshow(cloud)
    plt.axis('off')
    plt.show()
```

## Findings and Conclusions

Firstly, in all three campaigns, immigration features as a hot topic. We see tendencies of *renationalization*; "tendencies of creating ever new borders (and even walls), of linking the nation state and citizenship with nativist body politics" (Wodak 2015). There seems to be some form of social othering, as well as the construction of an **antagonism between us versus them** (Greven 2016). These have the effect of **mobilizing the fears and anxieties of the population**, swaying them in favour of conservative politics. As a result, a **politics of exclusion and fear** is legitimized by the right wing party.

Cowles (1994) offers a mode of inquiry - **ethnocentrism and multiculturalism - that is opposed to such reductive thinking**. He claims that "the threat of multiculturalism is an ipseity (focus on the self) coupled with the fear of the loss of power over the Stranger". It is useful to embrace the complexity and variation of human experiences, instead of forming a "us vs others" construct based on origin, race, religion etc.

A second feature of right-wing populism is the **juxtaposition of an elite (and often corrupt) political class against the rest of the populace** (Greven 2016). This featured most prominently in Trump's campaign in his criticism of the Clinton administration and his derogatory comments towards political rival Hillary Clinton. We also note that bread-and-butter issues like healthcare, housing, the economy and tax feature heavily. This is aligned with Marx's model of base and superstructure - that the "base of any society consists of the economic forces that regulate the consumption, production, and distribution of goods and services" (Cowles 1994). In right-wing ideology, political elites are deemed to be incognizant of on-the-ground concerns, and right-wing leaders promise economic reforms that will particularly appeal to the working class. As Greven (2016) suggests, "working class support of right-wing populism might at this point be more an expression of protest". Ekstrom et. al (2018) further supports this hypothesis.

*"... populist politicians appeal to ‘the people’ by claiming the moral high ground, subverting and undermining the establishment, and producing **identity performances that signify closeness to ‘the people’, as opposed to the perceived remoteness of mainstream political elites**. Thus, they claim social and cultural capital in their respective contexts of communication"*

## Limitations and Future Work

Because of time constraints, I could only work with 20 sources for the Trump campaign, 20 sources for the Brexit campaign and 10 sources for the Liberal-National campaign in Australia. Furthermore, I **did not have an objective and consistent method for selecting my data**. Further work could entail analyzing the full repository of speech transcripts for each of these campaigns. This would improve scientific objectivity and the comprehensiveness of the data.

I also realized that the **tools that we are taught are not perfect**. For instance:
    - the lemmatizer reduces "us" to "u"
    - the Tfidf Vectorizer does not handle contractions like "don't" and "haven't" very well
    - the stop word list is not comprehensive enough to handle words like "gonna"     
Given greater proficiency in Python, I would investigate these unexpected results and mitigate these issues.

Given more time and resources, it would be useful to **expand the scope of the project to study more campaigns**. I initially wanted to analyze Marine Le Pen's right-winged coalition in France, but found difficulty in locating credible translated transcripts of campaign speeches. It would also be interesting to **study the opposite camps of these right-wing campaigns** - do left-wing campaigns employ contrasting rhetorics and narratives? Furthermore, our analyses **fail to explain more microscopic voting patterns** (e.g. why are most Trump and Brexit voters middle class?).

Finally, in line with the ethos of the digital humanities, **computational text analysis and machine learning are supposed to supplement close reading, and not replace it**. In my interpretation, I had to make a lot of my own inferences. These inferences are likely clouded by my own biases and background knowledge. In order to fairly interpret the word clouds generated, there is a need to perform more close reading and cross-references to academic literature.
